```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/NLP/Korean_NLP/data/train.csv")
augmented_df = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/NLP/Korean_NLP/back_translation_data/augmented_corpus.csv")
```

```python
# 출처: 전상민 캠퍼님 코드 (https://github.com/boostcampaitech2/klue-level2-nlp-04/blob/JSM/description/EDA.md)
# back translation 하기 전 데이터 분포
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
ax.barh(sorted(df.label.unique(), reverse=True), df.label.value_counts().sort_index(ascending=False), color=sns.color_palette('hls', 30))
for idx, score in enumerate(df.label.value_counts().sort_index(ascending=False)):
    ax.text(s=score,x=score+50, y=idx, ha='left', va='center', fontweight='semibold')
    
plt.show()
```

```python
# back translation 하고난 후 데이터 분포
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
ax.barh(sorted(augmented_df.label.unique(), reverse=True), augmented_df.label.value_counts().sort_index(ascending=False), color=sns.color_palette('hls', 30))
for idx, score in enumerate(augmented_df.label.value_counts().sort_index(ascending=False)):
    ax.text(s=score,x=score+50, y=idx, ha='left', va='center', fontweight='semibold')
    
plt.show()
```
